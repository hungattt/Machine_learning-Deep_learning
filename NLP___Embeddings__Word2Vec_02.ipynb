```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Embedding, GlobalAveragePooling1D
```

```python
import tensorflow as tf

import tensorflow_datasets as tfds
imdb, info = tfds.load("imdb_reviews/subwords8k", with_info=True, as_supervised=True)
```

```python
train_data, test_data = imdb['train'], imdb['test']
```

```python
tokenizer = info.features['text'].encoder
```

```python
train_data = train_data.padded_batch(64, tf.compat.v1.data.get_output_shapes(train_data))
test_data = test_data.padded_batch(64,  tf.compat.v1.data.get_output_shapes(test_data))
```

```python
for item in train_data:
  print(item)
```

```python
# Xy dng model

# FIXME: To model Sequential

model = Sequential()

# FIXME: Thm mt lp Embedding

model.add(Embedding(tokenizer.vocab_size,64))

model.add(GlobalAveragePooling1D()) # Ta hon ton c th s dng mt lp Convolution 1D sau embedding  tm mi quan h gia cc t

# FIXME: Thm mt lp Dense, ch  c activation

model.add(Dense(6,activation='relu'))

# FIXME: Thm mt lp Dense cui, ch  c activation

model.add(Dense(1,activation='sigmoid'))
```

```python
num_epochs = 10

# FIXME: Compile Optimizer v Loss function

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])

# FIXME: Tin hnh training. Mc tiu: 87 %

history = model.fit(train_data,epochs=num_epochs,validation_data=test_data)
```
