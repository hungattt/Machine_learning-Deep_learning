```python
import cv2
import torch
import numpy as np
from models import build_model
from data_loader import get_test_loader
from config import config
from utils import draw_bbox

def visualize_attention(image_path, model_path, config):
    # Load the pre-trained model
    model = build_model(config)
    model.load_state_dict(torch.load(model_path, map_location=torch.device('cpu')))
    model.eval()

    # Load the image
    image = cv2.imread(image_path)
    h, w = image.shape[:2]

    # Prepare the input
    test_loader = get_test_loader(config)
    input_tensor, _ = test_loader.dataset.prepare_input(image)
    input_tensor = input_tensor.unsqueeze(0)

    # Get the attention map
    with torch.no_grad():
        _, _, attention_map = model(input_tensor)

    # Resize the attention map to match the image size
    attention_map = attention_map.squeeze().numpy()
    attention_map = cv2.resize(attention_map, (w, h), interpolation=cv2.INTER_LINEAR)

    # Normalize the attention map to [0, 255]
    attention_map = ((attention_map - attention_map.min()) / (attention_map.max() - attention_map.min()) * 255).astype(np.uint8)

    # Apply a color map to the attention map
    attention_map_color = cv2.applyColorMap(attention_map, cv2.COLORMAP_JET)

    # Overlay the attention map on the original image
    overlay = cv2.addWeighted(image, 0.5, attention_map_color, 0.5, 0)

    # Display the results
    cv2.imshow("Original Image", image)
    cv2.imshow("Attention Map", attention_map_color)
    cv2.imshow("Overlay", overlay)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    image_path = "path/to/your/input/image"
    model_path = "path/to/your/pretrained/model"
    visualize_attention(image_path, model_path, config)
```

```python
import cv2
import torch
import numpy as np
from models import create_DBNet
from data_loader import get_test_loader
from config import load_config
from utils import draw_bbox

def visualize_attention(image_path, model_path, config_file):
    # Load the config file
    config = load_config(config_file)

    # Load the pre-trained model
    model = create_DBNet(config)
    model.load_state_dict(torch.load(model_path, map_location=torch.device('cpu')))
    model.eval()

    # Load the image
    image = cv2.imread(image_path)
    h, w = image.shape[:2]

    # Prepare the input
    test_loader = get_test_loader(config)
    input_tensor, _ = test_loader.dataset.prepare_input(image)
    input_tensor = input_tensor.unsqueeze(0)

    # Get the attention map
    with torch.no_grad():
        _, _, attention_map = model(input_tensor)

    # Resize the attention map to match the image size
    attention_map = attention_map.squeeze().numpy()
    attention_map = cv2.resize(attention_map, (w, h), interpolation=cv2.INTER_LINEAR)

    # Normalize the attention map to [0, 255]
    attention_map = ((attention_map - attention_map.min()) / (attention_map.max() - attention_map.min()) * 255).astype(np.uint8)

    # Apply a color map to the attention map
    attention_map_color = cv2.applyColorMap(attention_map, cv2.COLORMAP_JET)

    # Overlay the attention map on the original image
    overlay = cv2.addWeighted(image, 0.5, attention_map_color, 0.5, 0)

    # Display the results
    cv2.imshow("Original Image", image)
    cv2.imshow("Attention Map", attention_map_color)
    cv2.imshow("Overlay", overlay)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

if __name__ == "__main__":
    image_path = "path/to/your/input/image"
    model_path = "path/to/your/pretrained/model/totaltext_resnet50"
    config_file = "config/totaltext_resnet50_deform_thre.yaml"
    visualize_attention(image_path, model_path, config_file)
```
